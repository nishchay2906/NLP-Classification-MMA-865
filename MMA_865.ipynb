### Part 1: What is “Big Data” and how is it different than “regular data”?

- Data that is very large and/or unstructured such as text, images etc., and traditional technologies cannot be used to handle their processing. Such type of data is referred to as Big Data. Big data keeps on increasing from time to time and can be inconsistent. Big Data is also a blanket term for techniques and technology stacks used to analyze big data, like Hadoop, spark,etc. Big data is high in volume, flows at a very high velocity, and is of different varieties.


- Regular data is structured data that is consistent and small in size and can be easily handled with traditional technologies such as RDBMS and does not flow at a very high velocity like streaming data, and the process and framework used to analyze Big data are different.


- Regular data can be stored and analyzed on a single machine, but big data is usually stored across multiple machines.


- For example, Data stored in Enterprise Data warehousing systems used by big companies is the regular data that we referred to; on the other hand, data such as streaming data or text review data which is high in volume, is an example of big data.

### Part 2: What is Hadoop? Hint: What problems in previous data storage and processing was Hadoop designed to solve? How did Hadoop accomplish that?

- As data became more complex, voluminous, and evergrowing, it was impossible to store and analyze such data, which resulted in a more efficient framework that was used to store and analyze such data, in 2006 by the emergence of Hadoop.


- Hadoop is a framework that allows users to store huge amounts of data across different machines and allows different machines to work together to process Big data. It can store enormous amounts of data in different varieties, such as text, images, or other unstructured data, such as JSON.


- Hadoop has four different components, namely
   - HDFS "Hadoop distributed file system" This is where all the data is stored
   - YARN - 'Yet another resource negotiator ' This is the component through which it is decided which machine would perform what job."
   - Mapreduce - "This is a programming model through which data is accessed through HDFS."
   - Hadoop Common - Resource and utilities for other Hadoop components
   
   
   
- Advantages of Hadoop 
   - Hadoop stores data in chunks and has backup available readily in multiple nodes so that even when one node fails, the processing still does not stop.
   - It is very cost-effective.
   - It is very scalable, and the number of machines used can be adjusted according to the requirements
   - It can support data from multiple sources.
   - Hadoop hides system-level details from the programmer


### Part 3: How does Big Data and the cloud help Machine Learning?

     - Memory Bound    
     
     - Computational Bound
     
     

- The more complex the model we use on a larger dataset, the more computation-heavy it becomes, and hence takes a long time to train and be used. Therefore, after a point, it becomes counterproductive to run such heavy computations to improve your model's performance by a fragment, and because of this, we can end up using a less reliable model. - With the emergence of Big Data and the cloud, These bounds on Machine learning have been pushed significantly, and now much larger datasets can be used to train the dataset, more computational power can be utilized according to the requirement, and now different types of data can also be easily analyzed. ### Part 4: What is NoSQL?

- NoSQL or "NOT ONLY SQL" is a way of storing and retrieving information in a non-traditional way which essentially means that the data is not stored in rows and columns. Non-relational databases store data according to the way that data is structured, e.g., key/value, JSON, XML, Graph, etc. Every item in NoSQL can have a different structure stored within a structure.



- Using NoSQL, we can store and manage Big data that can be structured, unstructured or semi-structured. Also, it is more efficient to read and write a large volume of data using NoSQL.



- In NoSQL databases, we can store various data types with flexible structures.



- NoSQL databases are faster and more scalable.



- In NoSQL databases do not usually follow the ACID properties, but instead, they follow the BASE properties, which (Basically Available, Soft state, Eventual consistency) means. 


     -  Basically Available -  Ensures the availability of Data
     -  Soft State - The state of the system can change over time
     -  Eventual consistency- The system will eventually become consistent once it stops receiving input.

## Feature Extraction

```python
# sentence length
df_train['Sentence_len'] = df_train.Sentence.apply(len)
print("Mean Sentence length: ", df_train['Sentence_len'].mean())
print("Min Sentence length: ", df_train['Sentence_len'].min())
print("Max Sentence length: ", df_train['Sentence_len'].max())
df_train.Sentence_len.plot.hist()



# sentiment compound score using SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent = SentimentIntensityAnalyzer()
def check_sentiment(x):
    return sent.polarity_scores(x)['compound']
df_train['compound_score'] = df_train['Sentence'].apply(lambda x: check_sentiment(x))
```

```python
df_train.head()
```

## Preprocessing

```python
#dropping irrelavant records
df_train = df_train.drop(df_train[df_train.Sentence == "#NAME?"].index).reset_index(drop = True)
```

```python
text = list(df_train['Sentence'])
```

```python
import re
from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

corpus = []

for i in range(len(text)):
    r = re.sub('[^a-zA-Z]', ' ', text[i]) #Regular expression
    r = r.lower() #lowering case
    r = r.split() # splitting
    r = [word for word in r if word not in stopwords.words('english')] #stopwords remover
    r = [lemmatizer.lemmatize(word) for word in r] #lemmatizer
    r = ' '.join(r)
    corpus.append(r)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

# fit transform the TfidfVectorizer
X = tfidf.fit_transform(corpus)

# convert X to array
X_array = X.toarray()

features = pd.DataFrame(X_array)
features.columns = tfidf.get_feature_names()

new_data_train = pd.concat([df_train,features], axis=1)
new_data_train.head()
```

```python
new_data_train = new_data_train.drop(['Sentence'], axis=1)
```

```python
X = new_data_train.drop(columns  = ["Polarity"])
y = new_data_train.Polarity
```

## Modeling

```python
lr = LogisticRegression()

scoring = {'acc': 'accuracy',
           'prec_macro': 'precision_macro',
           'rec_micro': 'recall_macro',
           'f1_macr':'f1_macro',
           'roc_auc':'roc_auc'}
scores = pd.DataFrame(cross_validate(lr, X=X, y=y,scoring=scoring, cv=10, n_jobs=-1))
```

```python
scores.test_f1_macr.sort_values()
```

```python
lr.fit(X,y)
```

```python
#hyper parameter tuning
hyper_params = {
        
        'class_weight':  ['balanced'],
        'solver': ['saga'],

        "penalty": ["l1",'l2','elasticnet'],

        "C": [100, 10, 1.0, 0.1, 0.01],
        

        "random_state": [42],
  }
    



search = GridSearchCV(lr, hyper_params, scoring='f1_macro', n_jobs=-1, cv=10)
# execute search
result = search.fit(X, y)
# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```
